##  1.3 样本数据的基本统计描述

即最简单的标准差，数量，平均值，范围，变异系数（标准差/平均值）等数值的计算。

```python
statistics = data.describe()  #保存基本统计量
statistics.loc['range'] = statistics.loc['max']-statistics.loc['min']  #极差
statistics.loc['var'] = statistics.loc['std']/statistics.loc['mean']  #变异系数
statistics.loc['dis'] = statistics.loc['75%']-statistics.loc['25%']  #四分位数间距
print(statistics)
```

# 2.异常值分析

异常值即数据中存在不合理数据需要进行剔除，即箱模型上下界的数据，

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False   

features = data.columns
nrows = (len(features) + 1) // 3
fig, axes = plt.subplots(nrows=nrows, ncols=3, figsize=(20, 6 * nrows), facecolor='white')
for i, feature in enumerate(features):
    row = i // 3  # 计算当前子图所在的行
    col = i % 3   # 计算当前子图所在的列
    ax = axes[row, col]  # 获取当前子图
    data.boxplot(column=feature, ax=ax)  # 绘制箱线图
    ax.set_title(f'{feature} 箱线图')  # 设置子图标题
if len(features) % 3 != 0:
    axes[-1, -1].axis('off')  # 隐藏最后一个子图
plt.tight_layout()  
plt.show() 
```

# 陕西省农村居民人均消费支出与收入水平对比 (2011-2020)

```python
plt.figure(figsize=(12, 8), facecolor='white')
plt.plot(data.index, data['陕西省农村居民人均消费支出'], label='人均消费支出', marker='o')
plt.plot(data.index, data['陕西省农村居民收入水平'], label='居民收入水平', marker='s')
plt.title('陕西省农村居民人均消费支出与收入水平对比 (2011-2020)', fontsize=16)
plt.xlabel('年份', fontsize=12)
plt.ylabel('元', fontsize=12)
for x, y in zip(data.index, data['陕西省农村居民人均消费支出']):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=10)
for x, y in zip(data.index, data['陕西省农村居民收入水平']):
    plt.text(x, y, f'{y:.0f}', ha='center', va='top', fontsize=10)
plt.grid(True)  # 添加网格线
plt.legend()   # 显示图例
plt.show()
```

# 陕西省农村居民人均消费支出与数字普惠金融总指数对比 (2011-2020)

# 财政支农与产业结构对比 (2011-2020)
