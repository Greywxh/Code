```python
traffic=data[['Management Type','participant']]
traffic_count=traffic.groupby("Management Type")["participant"].count()
traffic_count
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line
x=[str(i) for i in list(range(1,5))]
y=list(traffic_count)
line=(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name="participant",y_axis=y, is_smooth=True)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="不同类型农户借贷参与折线图"),
        yaxis_opts=opts.AxisOpts(
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),)
)
line.render_notebook()
```
